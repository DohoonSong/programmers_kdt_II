## I. Matplotlib 시작하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.plot([1,2,3,4,5])
#plt.show()
```

```python
plt.figure(figsize=(1,1)) # 도면의 선언

plt.plot([0,1,2,3,4])
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])

plt.plot(x,y)
```

```python
x = np.arange(-10,10,0.01) # start, end, interval

plt.xlabel('xvalue')
plt.ylabel('f(x)value')

plt.axis([-5, 5, 0, 25]) # xmin, xmax, ymin, ymax
plt.plot(x, x**2)
```

```python
x = np.arange(-10,10,0.01) # start, end, interval

plt.title('a graph of y=x^2')

plt.xlabel('xvalue')
plt.ylabel('f(x)value')

plt.axis([-5, 5, 0, 25]) # xmin, xmax, ymin, ymax

plt.xticks([i for i in range(-5,5,1)]) # x축의 tick을 지정.
plt.yticks([i for i in range(0,27,3)]) # y축의 tick을 지정
plt.plot(x, x**2, label='trend')

#plt.legend()
```

```python
x = np.arange(-10,10,0.01) # start, end, interval

plt.title('a graph of y=x^2')

plt.xlabel('xvalue')
plt.ylabel('f(x)value')

plt.axis([-5, 5, 0, 25]) # xmin, xmax, ymin, ymax

plt.xticks([i for i in range(-5,5,1)]) # x축의 tick을 지정.
plt.yticks([i for i in range(0,27,3)]) # y축의 tick을 지정
plt.plot(x, x**2, label='trend')

plt.legend()
```

```python
plt.show()
```

## II. Matplotlib Case Study

```python
x = np.arange(20)
y = np.random.randint(0,29,20)
plt.plot(x,y)
plt.axis([0,20,0,20])
plt.yticks([i for i in range(0,31, 5)])

plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot((x,y))
plt.title('Box plot of x / y')

"""
맨 위: single_max
박스 맨 위: Q3(75%)
박스 가운데 노란선: Q2(median, 50%)
박스 맨 아래: Q1(25%)
맨 아래: single_min
"""
plt.show()
```

```python
plt.bar(x,y)

plt.xticks([i for i in range(0,22,1)])
plt.show()
```

```python
plt.hist(y, bins=np.arange(0,29,2))
plt.xticks(np.arange(0,29,2))
plt.show()
```

```python
z = [100 * i for i in range(1,5,1)]

plt.pie(z, labels=['a','b','c','d'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(0,22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins = x)
plt.show()
```

```python
sns.kdeplot(y, shade = True)

plt.show()
```

```python
vote_df = pd.DataFrame({"name":['Andy', 'Bob', 'Cat'], "vote":[True, True, False]})

vote_count = vote_df.groupby("vote").count()

plt.bar(x=[False, True], height = vote_count['name'])

plt.show()
```

```python
sns.countplot(x=vote_df['vote'])

plt.show()
```

```python
covid = pd.read_csv('./country_wise_latest.csv')

s = sns.catplot(x='WHO Region', y='Confirmed', data=covid)# , kind='violin')
s.fig.set_size_inches(10, 3)
plt.show()
```

```python
s = sns.stripplot(x='WHO Region', y='Recovered', data=covid)

plt.show()
```

```python
s = sns.swarmplot(x='WHO Region', y='Recovered', data=covid)

plt.show()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
